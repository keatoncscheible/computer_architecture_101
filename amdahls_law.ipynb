# Amdahl's Law

At its core, Amdahl's Law addresses the impact of parallelization on the overall performance of a system. It helps us understand how much improvement in execution time we can achieve by applying parallel processing techniques to a program, given certain constraints.

The law is typically expressed using the following formula:

$\text{Speedup} = \frac{1}{(1 - P) + \frac{P}{N}}$

Where:
- (P) is the portion of a program that can be parallelized (expressed as a decimal, not a percentage).
- (N) is the number of processing units (cores, threads, etc.) available for parallel execution.

## Understanding the Formula:

1. **The Serial Fraction (1 - P):** This part of the formula represents the portion of the program that cannot be parallelized and must be executed sequentially. It's essentially the bottleneck that limits the potential speedup.

2. **The Parallel Fraction (P/N):** This term accounts for the portion of the program that can be parallelized. (P) represents the portion of the program that can be parallelized, and (N) is the number of processing units available for parallel execution. As (N) increases, this fraction contributes to the potential speedup.

## Key Insights

1. As the number of processing units (N) increases, the potential speedup initially improves. However, this improvement eventually saturates due to the serial fraction (1 - P) becoming more dominant. In other words, the program's sequential parts limit how much speedup can be achieved, regardless of how many processing units are available.

2. Amdahl's Law emphasizes the importance of identifying and optimizing the serial fraction in a program. Even a small reduction in the serial portion can lead to substantial improvements in overall performance.

3. The law highlights the diminishing returns of adding more processing units to a parallel system. If the serial fraction is relatively high, the additional processing units might not provide significant speedup.

## Real-World Application

Amdahl's Law is valuable for guiding design decisions in parallel computing systems. It underscores the need for a balanced approach to parallelization, focusing on both parallelizing code segments and optimizing sequential sections to achieve substantial speedup.

In conclusion, Amdahl's Law serves as a cautionary reminder that while parallel processing can greatly enhance performance, there are limits imposed by the nature of the program itself. It encourages architects and programmers to carefully analyze the balance between parallel and sequential execution when designing efficient computing systems.

```python
import numpy as np
import matplotlib.pyplot as plt

# Proportions of the program that are parallizable
P_arr = [0.1, 0.5, 0.75, 0.9, 0.95]

# Number of Processors
N_arr = np.arange(1, 1000, 1)

# Ahmdal's Law Equation
speedup = lambda P, N: 1 / ((1 - P) + P/N)

# Plot Speedup vs Number of Processors
for P in P_arr:
    plt.plot(N_arr, speedup(P, N_arr))

# Configure plot
plt.xscale('log')
plt.grid()
plt.title("Amdahl's Law")
plt.xlabel("Number of Processors")
plt.ylabel("Speedup")
plt.legend([f"P = {P}" for P in P_arr])
pass
```
